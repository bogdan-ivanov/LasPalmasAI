```python
from nltk.corpus import twitter_samples
```

```python
negative = []

for tweet in twitter_samples.strings('negative_tweets.json'):
    negative.append(tweet)
    
```

```python
negative[:15]
```

```python
positive = []

for tweet in twitter_samples.strings('positive_tweets.json'):
    positive.append(tweet)
```

```python
positive[:15]
```

```python
import random

def build_corpus():
    labelled_negative = [(tweet, '-') for tweet in negative]
    labelled_positive = [(tweet, '+') for tweet in positive]
    corpus = labelled_negative + labelled_positive
    random.shuffle(corpus)
    
    return corpus
```

```python
corpus = build_corpus()
```

```python
corpus[:10]
```

```python
len(corpus)
```

```python
training_corpus = corpus[:8000]
test_corpus = corpus[8000:]
```

```python
len(training_corpus)
len(test_corpus)
```

```python
import string
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
print(stop_words)


def preprocess(text):
    words = word_tokenize(text.lower())
    
    words = [w for w in words if w not in stop_words and w not in list(string.punctuation)]
    
    stems = [stemmer.stem(w) for w in words]
    
    return {s: True for s in stems}

```

```python
stopwords.words('malayesian')
```

```python
preprocess("I bought apples and oranges")
```

```python
print(stemmer.stem('oranges'))
print(stemmer.stem('orange'))
```

```python
preprocessed_training_corpus = [(preprocess(tweet), label) for tweet, label in training_corpus]
preprocessed_test_corpus = [(preprocess(tweet), label) for tweet, label in test_corpus]
```

```python
preprocessed_training_corpus[:5]
```

```python
from nltk.classify import NaiveBayesClassifier
```

```python
clf = NaiveBayesClassifier.train(preprocessed_training_corpus)
```

```python
print(clf.classify(preprocess("This movie was veeeery bad!")))
print(clf.classify(preprocess("Awesome game today!")))
print(clf.classify(preprocess("The meeting turned out as expected")))
print(clf.classify(preprocess("That's cool")))
print(clf.classify(preprocess("yeah, sure, he's suuuuper smart!")))
print(clf.classify(preprocess("You're an average developer!")))
print(clf.classify(preprocess("This was not so good :(")))
```

```python
from nltk.classify.util import accuracy
print(accuracy(clf, preprocessed_test_corpus))
```

```python
clf.show_most_informative_features(1000)
```

```python
from nltk.stem import WordNetLemmatizer

lmtz = WordNetLemmatizer()
print(lmtz.lemmatize('running', 'v'))
print(lmtz.lemmatize('running', 'n'))

from nltk import pos_tag

sent = "I bought apples and oranges".split()

print(pos_tag(sent))

lemmas = []
for pair in pos_tag(sent):
    word, tag = pair
    if tag[0] == 'N':
        lemmas.append(lmtz.lemmatize(word, 'n'))
    elif tag[0] == 'V':
        lemmas.append(lmtz.lemmatize(word, 'v'))
    else:
        lemmas.append(word)
        
print(lemmas)

lmtz.lemmatize('running', 'n')
```
